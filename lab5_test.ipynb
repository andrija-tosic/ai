```python
from collections import deque
from lab5 import solve, print_solution
from pprint import pprint

N = 8

availability_graph = {
    i: { x for x in range(N) } for i in range(N)
}

# pprint(free_graph)
visited_rows = deque([], N)
visited_cols = deque([], N)

res = solve(N, 2, availability_graph, visited_rows, visited_cols)

print_solution(N, res)

print("Result (row, col):")
pprint(res)
```

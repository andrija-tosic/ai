```python
import time
from lab4 import Point2D, diag_wall, wall, board_to_graph, search, draw_board
```

```python
t0 = time.time()

n = 4
start = Point2D(2, 1)
end = Point2D(1, 2)
forbidden = {Point2D(1, 1), Point2D(2, 3)}

(graph, start_key, end_key) = board_to_graph(n, start, end, forbidden)

path, iterations = search(graph, start_key, end_key)


# pprint(graph)

# print(path)
t1 = time.time()


total = (t1-t0)*1000

draw_board(n, start, end, forbidden, path)
print(f"{iterations} iteracija.")
print(f"Duzina puta: {len(path)}.")
print(f"Time: {total} ms.")
```
